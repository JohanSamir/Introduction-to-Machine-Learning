```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Regression models are used to predict target variables on a continuous scale.
#https://archive.ics.uci.edu/ml/machine-learning-databases/housing/
# sep : str, default ‘,’ Delimiter to use.


#Regular Expressions Cheat Sheet by DaveChild
#https://www.cheatography.com/davechild/cheat-sheets/regular-expressions/
df = pd.read_csv('/home/johan/repos/GitHub/Introduction-to-Machine-Learning/Datasets/housing.data',header=None, sep='\s+')
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS','NOX', 'RM', 'AGE', 'DIS', 'RAD','TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.head()
```

```python
sns.set(style='whitegrid', context='notebook')

#Exploratory Data Analysis (EDA) is an important and recommended first step prior
#to the training of a machine learning mode

cols = ['LSTAT', 'INDUS', 'NOX', 'RM', 'MEDV']
#To plot the scatterplot matrix, we will use the pairplot function from the seaborn library
#( http://stanford.edu/~mwaskom/software/seaborn/ ), which is a Python library
#for drawing statistical plots based on matplotlib:
sns.pairplot(df[cols], size=2.5);
plt.show()
```
